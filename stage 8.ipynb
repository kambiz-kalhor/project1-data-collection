```python
####################### extracting accession numbers ########################
print('making list of accession numbers...............')

list_of_accession_numbers = []
for row in range(0,len(my_dataframe)):
    for column in range (2,300,7):
        accession_number = my_dataframe.iloc[row,column]
        if str(accession_number) != 'nan':
            list_of_accession_numbers.append(accession_number)
```

```python
len(list_of_accession_numbers)
```

```python
################################################################################
#now its time to extract the subsequent sequence data using their accession numbers
erorrs = []
list_id = []
list_seq =[]
list_description =[]
for i in list_of_accession_numbers:
    print('searching for', i)
    try:
        handle = Entrez.efetch(db = "nucleotide", id = i, rettype = "fasta")
        record = SeqIO.read( handle, "fasta" )
        list_id.append(record.id)
        list_description.append(record.description)
        list_seq.append(record.seq)
        print('sequence found' , len(list_seq))
    except:
        erorrs.append(i)
        print('without any sequnce on NCBI>nucleotide: ', len(erorrs))


list_of_tuples = list(zip(list_description, list_seq))
df_seq = pd.DataFrame(list_of_tuples,columns = ["description",'sequence'])
df_seq.to_csv(first_output_path)
print('total errors: ', len(erorrs))
```

```python


#############################################################################
# searching to find sequnces for errors
# searching assembly database in NCBI
print('downloading from assembly>NCBI   please wait.....................')
link_list = []
def get_assembly_summary(id):
    """Get esummary for an entrez id"""
    esummary_handle = Entrez.esummary(db="assembly", id=id, report="full")
    esummary_record = Entrez.read(esummary_handle)
    return esummary_record


def get_links(accesstion_numbers):
    """ Get links from internet and return list of links."""
    global link_list
    # read links from interent
    for item in accesstion_numbers:
        print(item)
        links = get_assemblies_link_from_accession_number(item)
        print(links)
        link_list.extend(links)

    return link_list


def get_assemblies_link_from_accession_number(term):
    """Download genbank assemblies for a given search term.
    Args:
        term: search term, usually organism name
    """

    # provide your own mail here # I wrote the email at the begining of the codes
    handle = Entrez.esearch(db="assembly", term=term, retmax="200")
    record = Entrez.read(handle)
    ids = record["IdList"]
    links = []
    for aid in ids:
        summary = get_assembly_summary(aid)  # get summary
        url = summary["DocumentSummarySet"]["DocumentSummary"][0]["FtpPath_RefSeq"]
        if url == "":
            continue
        label = os.path.basename(url)
        # get the fasta link - change this to get other formats
        link = url + "/" + label + "_genomic.fna.gz"
        link = link.replace("ftp://", "https://")
        links.append(link)
    return links


def write_links(links, filename):
    """ Get list of links and write them to a text file"""
    with open(filename, "wb") as f:
        pickle.dump(links, f)


def read_links(filename):
    with open(filename, "rb") as f:
        links = pickle.load(f)
    return links


def download(url, path):
    """Download file based on url and save file to the path"""
    response = requests.get(url)

    if response.ok:
        print("response is ok file is downloading ... ")
        # start to download file from url.
        with open(path, "wb") as f:
            f.write(response.content)
    else:
        print("Error!", response.status_code)
        return False

    print("File downloaded succusfully.")
    return True


def main():
    # giving the accesstion_numbers
    accesstion_numbers = erorrs

    # read links from internet
    links = get_links(accesstion_numbers)
    # write links to a pickle file
    write_links(links, "links.pickle")

    # read links from a pickle file
    links = read_links("links.pickle")
    print(links)

    # download of each link's data and save them with path name
    for link in links:
        filename = link.split("/")[-1]
        download(link, f"files\\{filename}")
    # download data from each links


if __name__ == "__main__":
    main()
```
